# Sketch data processing

## Load data

```python
import pandas as pd
df = pd.read_csv('../../data/housing.csv')
df
```

## Filter data

### Masking

```python
df['ocean_proximity'].unique()
mask = df['ocean_proximity'] == 'NEAR OCEAN'
dfr = df[mask]
dfr
```

### Visualization

```python
import plotly.express as px
px.scatter(dfr, x='median_house_value', y='households', size='housing_median_age', color="population", size_max=15, height=700)
```
